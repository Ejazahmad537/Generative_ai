#Sentiment Analysis

```python
import transformers
from transformers import pipeline

pipe = pipeline("text-classification")
pipe("This movie is very boring")
```

```python
import transformers
from transformers import pipeline


pipe = pipeline(model="roberta-large-mnli")
pipe("I do not like this movie")
```

#We can also pass the string in the form of list

```python
#pipline for text classification
import transformers
from transformers import pipeline
pipe = pipeline("sentiment-analysis")
pipe(["This restaurant is awesome", "This restaurant is awful"])
```

#Text Summarization
we can summarize the text using the hugging face pipline for that we need to pass the pipeline the parameter as "summarizations'

```python
import transformers
from transformers import pipeline
```

```python
# text to be summarized
input_text = "Start by providing your text input. It could be a sentence or a paragraph."
print(input_text)
```

```python
#use bart in pytorch
summarizer = pipeline("summarization")
summarizer(f"Text_input: {input_text}", min_length=5, max_length=30)
```
